## Exemple avec le style *pyplot*

```python
x = ["JAN", "FEV", "MAR", "AVR"]
y1 = [4, 7, 6, 8]
y2 = [7, 5, 6, 9]

plt.figure(figsize=(10, 5), layout="constrained")

plt.plot(x, y1, marker="o", linestyle="dotted", label="Hommes")
plt.plot(x, y2, marker="o", label="Femmes")

plt.title("Un joli graphique")
plt.xlabel("Ann√©e 2020")
plt.ylabel("Nombre")
plt.legend();
```

## Exemple avec le style *object-oriented*

```python
x = ["JAN", "FEV", "MAR", "AVR"]
y1 = [4, 7, 6, 8]
y2 = [7, 5, 6, 9]

fig, ax = plt.subplots(figsize=(10, 5), layout="constrained")

ax.plot(x, y1, marker="o", linestyle="dotted", label="Hommes")
ax.plot(x, y2, marker="o", label="Femmes")
ax.set_title("Un joli graphique")
ax.set_xlabel("Ann√©e 2020")
ax.set_ylabel("Nombre")
ax.legend();
```

## Un graphique en barres

```python
x = ["JAN", "FEV", "MAR", "AVR"]
y = [4, 7, 6, 8]
plt.title("Nombre de jours de soleil par mois")
plt.bar(x, y, width=0.3);
```

## Graphique en secteurs

```python
labels = "0-19", "20-39", "44-59", "60-79", "80-99"
numbers = [15, 22, 25, 15, 7]
plt.title("R√©partition en fonction de l'√¢ge")
plt.pie(numbers, labels=labels, autopct="%.2f", startangle=90, counterclock=False);
```

## Plusieurs axes dans la m√™me figure

```python
x = ["JAN", "FEV", "MAR", "AVR"]
y1 = [4, 7, 6, 8]
y2 = [7, 5, 6, 9]

fig, axs = plt.subplots(1, 2, figsize=(14, 4), layout="constrained")

axs[0].plot(x, y1, marker="o", linestyle="dotted", label="Hommes")
axs[1].plot(x, y2, marker="o", label="Femmes");
```
