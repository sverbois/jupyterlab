```python
import matplotlib
from matplotlib import pyplot as plt
```

```python
print(matplotlib.__version__)
```

```python
x = ["JAN", "FEV", "MAR", "AVR"]
y1 = [4, 7, 6, 8]
y2 = [7, 5, 6, 9]

plt.plot(x, y1, marker='o', linestyle = 'dotted', label="Hommes")
plt.plot(x, y2, marker='o', label="Femmes")

plt.title("Un joli graphique")
plt.xlabel("Année 2020")
plt.ylabel("Nombre")
plt.legend();
```

```python
x = ["JAN", "FEV", "MAR", "AVR"]
y = [4, 7, 6, 8]
plt.title("Nombre de jours de soleil par mois")
plt.bar(x, y, width = 0.3);
```

```python
labels = '0-19', '20-39', '44-59', '60-79', '80-99'
numbers = [15, 22, 25, 15, 7]
plt.title("Répartition en fonction de l'âge")
plt.pie(numbers, labels=labels, autopct='%.2f', startangle=90, counterclock=False);
```
